```python
import matplotlib.pyplot as plt
```

```python
methods = ["DSB", "RF"]
pairs = ["gaussian_swiss_roll", "circles_swiss_roll", "moons_swiss_roll", "s_curve_swiss_roll"]
```

```python
f, axs = plt.subplots(len(methods), len(pairs), figsize=(12, 6))

fontsize = 10

plt.rcParams["font.family"] = "Arial"  # or "Arial", "Comic Sans MS", etc.

for ii, method in enumerate(methods):
    for jj, pair in enumerate(pairs):
        img_name = "{}_traj_{}.png".format(method, pair)
        img = plt.imread(img_name)
        axs[ii, jj].imshow(img)
        
        # axs[ii, jj].axis("off")
        axs[ii, jj].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
        for spine in axs[ii, jj].spines.values():
            spine.set_visible(False)
            
        title = "prior: {}".format(pair[:-len("_swiss_roll")])
            
        axs[0, jj].set_title(title, fontsize=fontsize)
        
    axs[ii, 0].set_ylabel(method, fontsize=fontsize)

f.subplots_adjust(
    left=0.02, right=1, top=1, bottom=0,
    hspace=0.01, wspace=0.1, 
)

plt.savefig("compare_traj", dpi = 500)
plt.show()
```
