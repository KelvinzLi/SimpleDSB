```python
import matplotlib.pyplot as plt
```

```python
dsb_img = plt.imread("DSB_traj.png")
rf_img = plt.imread("RF_traj.png")

dsb_img = dsb_img[:, 160: -160]
rf_img = rf_img[:, 160: -160]
```

```python
(1280 - 960) / 2
```

```python
f, axs = plt.subplots(1, 2)

fontsize = 10

axs[0].imshow(dsb_img)
axs[0].set_title("DSB Trajectory", fontsize = fontsize)
axs[0].axis('off')

axs[1].imshow(rf_img)
axs[1].set_title("RF Trajectory", fontsize = fontsize)
axs[1].axis('off')

f.subplots_adjust(wspace=0, hspace=0)

plt.tight_layout(pad = 0)

plt.savefig("compare_traj", dpi = 500)
plt.show()
```
